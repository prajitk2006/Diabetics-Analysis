Importing Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

File Input

```python
df = pd.read_csv('../data/diabetes.csv')
```

Cleaning Datasets with plotting

```python
cols_with_zeros = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
print(df[cols_with_zeros].describe())

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(df['Glucose'], ax=axes[0])
axes[0].set_title('Distribution of Glucose (Note the 0 spike)')
sns.histplot(df['BMI'], ax=axes[1])
axes[1].set_title('Distribution of BMI (Note the 0 spike)')
plt.show()
```
